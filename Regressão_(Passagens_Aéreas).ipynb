```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
file_path = "/content/Passagens aereas.xlsx"
df = pd.read_excel(file_path)

# Separando as variáveis independentes (X) e dependentes (y)
X = df[['Preço']]
y = df['Quantidade']

# Instanciando o modelo de regressão linear
model = LinearRegression()

# Ajustando o modelo aos dados
model.fit(X, y)

# Coeficiente angular (inclinação da reta)
coef_angular = model.coef_[0]

# Intercepto (ponto onde a reta corta o eixo y)
intercepto = model.intercept_

# Equação da reta (modelo linear)
equacao_reta = f'Quantidade = {coef_angular:.4f} * Preço + {intercepto:.4f}'

# Plotando o gráfico de dispersão com a linha de tendência
plt.figure(figsize=(8, 6))
plt.scatter(df['Preço'], df['Quantidade'], color='blue', label='Dados')
plt.plot(X, model.predict(X), color='green', label='Linha de tendência')
plt.title('Regressão Linear - Preço x Quantidade')
plt.xlabel('Preço')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)
plt.show()

# Equação de receita (Q * P)
equacao_receita = f'Receita = Preço * Quantidade'

# Calculando a receita para cada ponto de preço
df['Receita'] = df['Preço'] * df['Quantidade']

# Plotando o gráfico de receita em função do preço
plt.figure(figsize=(8, 6))
plt.plot(df['Preço'], df['Receita'], marker='o', linestyle='-', color='red')
plt.title('Receita em função do Preço')
plt.xlabel('Preço')
plt.ylabel('Receita')
plt.grid(True)
plt.show()

# Encontrando o preço ótimo (onde a receita é máxima)
preco_otimo = df['Preço'][df['Receita'].idxmax()]
receita_maxima = df['Receita'].max()

print(f'O preço ótimo para maximizar a receita é: R${preco_otimo:.2f}')
print(f'A receita máxima obtida é: R${receita_maxima:.2f}')
```
