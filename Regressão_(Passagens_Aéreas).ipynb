```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
file_path = "/content/Passagens aereas.xlsx"
df = pd.read_excel(file_path)

# Separando as variveis independentes (X) e dependentes (y)
X = df[['Preo']]
y = df['Quantidade']

# Instanciando o modelo de regresso linear
model = LinearRegression()

# Ajustando o modelo aos dados
model.fit(X, y)

# Coeficiente angular (inclinao da reta)
coef_angular = model.coef_[0]

# Intercepto (ponto onde a reta corta o eixo y)
intercepto = model.intercept_

# Equao da reta (modelo linear)
equacao_reta = f'Quantidade = {coef_angular:.4f} * Preo + {intercepto:.4f}'

# Plotando o grfico de disperso com a linha de tendncia
plt.figure(figsize=(8, 6))
plt.scatter(df['Preo'], df['Quantidade'], color='blue', label='Dados')
plt.plot(X, model.predict(X), color='green', label='Linha de tendncia')
plt.title('Regresso Linear - Preo x Quantidade')
plt.xlabel('Preo')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)
plt.show()

# Equao de receita (Q * P)
equacao_receita = f'Receita = Preo * Quantidade'

# Calculando a receita para cada ponto de preo
df['Receita'] = df['Preo'] * df['Quantidade']

# Plotando o grfico de receita em funo do preo
plt.figure(figsize=(8, 6))
plt.plot(df['Preo'], df['Receita'], marker='o', linestyle='-', color='red')
plt.title('Receita em funo do Preo')
plt.xlabel('Preo')
plt.ylabel('Receita')
plt.grid(True)
plt.show()

# Encontrando o preo timo (onde a receita  mxima)
preco_otimo = df['Preo'][df['Receita'].idxmax()]
receita_maxima = df['Receita'].max()

print(f'O preo timo para maximizar a receita : R${preco_otimo:.2f}')
print(f'A receita mxima obtida : R${receita_maxima:.2f}')
```
