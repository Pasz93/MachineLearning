```python
import pandas as pd
import requests

# Faz uma requisição para obter os dados da página da Wikipedia
requisicao = requests.get("https://pt.wikipedia.org/wiki/La_Liga_de_2023%E2%80%9324")

# Extrai todas as tabelas presentes na página e armazena em 'tabelas'
tabelas = pd.read_html(requisicao.text)

# Seleciona as tabelas de classificação e de jogos
tabela_classificacao = tabelas[8]
tabela_jogos = tabelas[11]

# Exibe as tabelas de classificação e de jogos
display(tabela_classificacao)
display(tabela_jogos)
```

```python
# Cria um mapeamento de apelidos para nomes completos dos times
nomes_times = list(tabela_jogos["Casa \ Fora"])
apelidos = list(tabela_jogos.columns)
apelidos.pop(0)
de_para_times = dict(zip(apelidos,nomes_times))

# Imprime o mapeamento dos apelidos para nomes completos
print (de_para_times)
```

```python
# Ajusta a estrutura da tabela de jogos para separar os nomes dos times e o placar
tabela_jogos_ajustada = tabela_jogos.set_index("Casa \ Fora")
tabela_jogos_ajustada = tabela_jogos_ajustada.unstack().reset_index()
tabela_jogos_ajustada = tabela_jogos_ajustada.rename(columns={"level_0": "Fora", "Casa \ Fora": "Casa",0: "Placar"})

# Define função para ajustar apelidos dos times nos nomes completos
def ajustar_apelido_time(linha):
  apelidos=linha["Fora"]
  nome = de_para_times[apelidos]
  return nome

# Aplica a função para ajustar os apelidos dos times na tabela de jogos
tabela_jogos_ajustada["Fora"] = tabela_jogos_ajustada.apply(ajustar_apelido_time, axis=1)

# Filtra os jogos onde o time da casa é diferente do time visitante
tabela_jogos_ajustada = tabela_jogos_ajustada[tabela_jogos_ajustada["Fora"]!=tabela_jogos_ajustada["Casa"]]

# Exibe os jogos futuros
display(tabela_jogos_ajustada)
```

```python
# Preenche os jogos ainda não realizados com "a jogar"
tabela_jogos_ajustada["Placar"] = tabela_jogos_ajustada["Placar"].fillna("a jogar")

# Separa os jogos já realizados dos jogos futuros
tabela_jogos_realizados = tabela_jogos_ajustada[tabela_jogos_ajustada["Placar"].str.contains("–")]
tabela_jogos_faltantes = tabela_jogos_ajustada[~tabela_jogos_ajustada["Placar"].str.contains("–")]
tabela_jogos_faltantes = tabela_jogos_faltantes.drop(columns=["Placar"])

# Exibe os jogos futuros
display(tabela_jogos_faltantes)
```

```python
# Separa os gols feitos em casa e fora nos jogos realizados
colunas = ["gols_casa", "gols_fora"]
tabela_jogos_realizados[colunas] = tabela_jogos_realizados["Placar"].str.split("–", expand=True)
tabela_jogos_realizados = tabela_jogos_realizados.drop(columns=["Placar"])
tabela_jogos_realizados["gols_casa"] = tabela_jogos_realizados["gols_casa"].astype(int)
tabela_jogos_realizados["gols_fora"] = tabela_jogos_realizados["gols_fora"].astype(int)

# Exibe os gols feitos em jogos já realizados
display(tabela_jogos_realizados)
```

```python

# Calcula as médias de gols feitos e sofridos em casa e fora para cada time
media_gols_casa = tabela_jogos_realizados.groupby("Casa").mean(numeric_only=True)
media_gols_casa = media_gols_casa.rename(columns={"gols_casa": "gols_feitos_casa", "gols_fora": "gols_sofridos_casa"})
media_gols_fora = tabela_jogos_realizados.groupby("Fora").mean(numeric_only=True)
media_gols_fora = media_gols_fora.rename(columns={"gols_casa": "gols_sofridos_fora", "gols_fora": "gols_feitos_fora"})

```

```python
# Cria a tabela de estatísticas com as médias de gols
tabela_estatisticas = media_gols_casa.merge(media_gols_fora, left_index=True, right_index=True)
tabela_estatisticas = tabela_estatisticas.reset_index()
tabela_estatisticas = tabela_estatisticas.rename(columns={"Casa": "time"})
display(tabela_estatisticas)
```

```python
from scipy.stats import poisson

# Define a função para calcular as pontuações esperadas para os jogos futuros
def calcular_pontuacao_esperada (linha):
  time_casa = linha["Casa"]
  time_fora = linha["Fora"]

  lambda_casa = tabela_estatisticas.loc[tabela_estatisticas["time"] == time_casa, "gols_feitos_casa"].iloc[0] * tabela_estatisticas.loc[tabela_estatisticas["time"] == time_fora, "gols_sofridos_fora"].iloc[0]

  lambda_fora = tabela_estatisticas.loc[tabela_estatisticas["time"] == time_fora, "gols_feitos_fora"].iloc[0] * tabela_estatisticas.loc[tabela_estatisticas["time"] == time_casa, "gols_sofridos_casa"].iloc[0]

  pv_casa = 0
  p_empate = 0
  pv_fora = 0

# Calcula as probabilidades de resultados possíveis para cada jogo futuro
  for gols_casa in range(8):
      for gols_fora in range(8):
          probabilidade_resultado = poisson.pmf(gols_casa, lambda_casa) * poisson.pmf(gols_fora, lambda_fora)
          if gols_casa == gols_fora:
            p_empate += probabilidade_resultado
          elif gols_casa > gols_fora:
            pv_casa += probabilidade_resultado
          elif gols_casa <  gols_fora:
            pv_fora += probabilidade_resultado

# Calcula as pontuações esperadas para cada time
  valor_esperado_casa = pv_casa * 3 + p_empate
  valor_esperado_fora = pv_fora * 3 + p_empate
  linha["pontos_casa"] = valor_esperado_casa
  linha["pontos_fora"] = valor_esperado_fora
  return linha


#print(pv_casa)
#print(p_empate)
#print(pv_fora)
```

```python
# Aplica o cálculo das pontuações esperadas para os jogos futuros
tabela_jogos_faltantes = tabela_jogos_faltantes.apply(calcular_pontuacao_esperada, axis=1)
display(tabela_jogos_faltantes)
```

```python
# Prepara a tabela de classificação atualizada
tabela_classificacao_atualizada = tabela_classificacao[["Equipe", "Pts"]]
tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada["Pts"].astype(int)
display(tabela_classificacao_atualizada)
```

```python
# Calcula as pontuações esperadas dos jogos futuros para cada time
pontuacao_casa = tabela_jogos_faltantes.groupby("Casa").sum(numeric_only=True)[["pontos_casa"]]
pontuacao_fora = tabela_jogos_faltantes.groupby("Fora").sum(numeric_only=True)[["pontos_fora"]]
display(pontuacao_casa, pontuacao_fora)
```

```python
# Atualiza a pontuação prevista na tabela de classificação
def atualizar_pontuacao_previsao(linha):
  equipe = linha["Equipe"]
  pontuacao = int(linha["Pts"]) + float(pontuacao_casa.loc[equipe, "pontos_casa"]) + float(pontuacao_fora.loc[equipe,"pontos_fora" ])
  return pontuacao

tabela_classificacao_atualizada ["Pts"] = tabela_classificacao_atualizada.apply(atualizar_pontuacao_previsao, axis=1)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.sort_values(by= "Pts",ascending=False)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.reset_index(drop=True)
tabela_classificacao_atualizada.index = tabela_classificacao_atualizada.index + 1

# Exibe a tabela de classificação atualizada
display(tabela_classificacao_atualizada)
```

```python
# Calcular probabilidades de vitória, derrota e empate para os jogos faltantes
def calcular_probabilidades(linha):
    pontos_casa = linha["pontos_casa"]
    pontos_fora = linha["pontos_fora"]

    total_pontos = pontos_casa + pontos_fora
    probabilidade_vitoria_casa = round(pontos_casa / total_pontos, 2)
    probabilidade_empate = round((3 - pontos_casa - pontos_fora) / total_pontos, 2)
    probabilidade_vitoria_fora = round(pontos_fora / total_pontos, 2)

    linha["prob_vitoria_casa"] = probabilidade_vitoria_casa
    linha["prob_empate"] = probabilidade_empate
    linha["prob_vitoria_fora"] = probabilidade_vitoria_fora
    linha["Time_Casa"] = linha["Casa"]
    linha["Time_Fora"] = linha["Fora"]
    return linha

# Aplicar o cálculo das probabilidades nos jogos faltantes
tabela_jogos_faltantes_com_prob = tabela_jogos_faltantes.apply(calcular_probabilidades, axis=1)

# Exibir os jogos faltantes com as probabilidades calculadas e os respectivos times
display(tabela_jogos_faltantes_com_prob[["Time_Casa", "Time_Fora", "prob_vitoria_casa", "prob_empate", "prob_vitoria_fora"]])
```

```python
# Selecionar apenas as colunas desejadas
colunas_selecionadas = tabela_jogos_faltantes_com_prob[["Time_Casa", "Time_Fora", "prob_vitoria_casa", "prob_empate", "prob_vitoria_fora"]]

# Salvar as colunas selecionadas em um arquivo Excel
nome_arquivo = 'tabela_jogos_faltantes_com_prob_la_liga.xlsx'
colunas_selecionadas.to_excel(nome_arquivo, index=False)
```
